```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
#data frame
df_conta = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

df_conta
```

```python
#Final do df - calda - padrÃ£o 5
df_conta.tail()
```

```python
# groupby == > Agrupamento do DataFrame  | agg

df_conta_agrupamento_dia = df_conta.groupby('day', sort = False).agg({'total_bill': np.sum}).reset_index()
df_conta_agrupamento_dia
```

```python
px.bar(df_conta_agrupamento_dia, x='day', y='total_bill')
```

```python
#FaÃ§a o agrupamento pelo o dia e  sexo
#1Â° DataFrame
df_conta_agrupamento_dia_sexo = df_conta.groupby(['day', 'sex'], sort=False).agg({'total_bill': np.sum}).reset_index()
df_conta_agrupamento_dia_sexo
```

```python
#FaÃ§a o agrupamento pelo o dia e  sexo
#2Â° GrÃ¡fico
px.bar(df_conta_agrupamento_dia_sexo, x='day', y='total_bill', color = 'sex')
```

```python
px.bar(df_conta_agrupamento_dia_sexo, x='day', y='total_bill', color = 'sex', barmode='group')
```

```python
#FaÃ§a o agrupamento pelo o dia e  sexo - barra deitada

px.bar(df_conta_agrupamento_dia_sexo, x='total_bill', y='day', color = 'sex', barmode='group')
```

```python
# FaÃ§a um grÃ¡fico de dispersÃ£o
px.scatter(df_conta, x='total_bill', y='tip', color='day', hover_name='time')
```

## Os mais comuns

A biblioteca __Pandas__ possui muitas funÃ§Ãµes e mÃ©todos implementados que nos ajudam, e muito, em nosso dia-a-dia na anÃ¡lise exploratÃ³ria de dados.

Nesta aula, vamos aprender a utilizar os mais comuns conforme estÃ£o listados abaixo:

* `rename` : renomeia os 'rÃ³tulos' dos eixos (Ã­ndices ou colunas)

* `apply` : aplica uma determinada funÃ§Ã£o ao longo de um eixo quando for chamado diretamente pelo _DataFrame_ ou aplica uma funÃ§Ã£o numa _Series_ em questÃ£o

* `map` : "mapeia" uma _Series_ aplicando determinada funÃ§Ã£o a cada elemento desta

* `agg` : agrega uma ou mais operaÃ§Ãµes Ã  uma _Series_

* `min` : retorna o valor mÃ­nimo de uma _Series_

* `max` : retorna o valor mÃ¡ximo de uma _Series_

* `sum` : retorna a soma dos valores de uma _Series_

* `mean` : retorna a mÃ©dia dos valores de uma _Series_

* `count` : conta as cÃ©lulas nÃ£o vazias de cada linha ou coluna

* `value_counts` : retorna uma _Series_ contendo a contagem de linhas exclusivas no _DataFrame_

* `sort_values` : classifica / ordena os valores ao longo de qualquer eixo

* `insert` : insire uma nova coluna no _DataFrame_ no local especificado

* `drop` : remova linhas ou colunas especificando diretamente seus nomes

ConstruÃ§Ã£o de novo DataFrame, mas desta vez, passando os parÃ¢metros nomeadas - data, columns e index. Repare que o parÃ¢metro nomeado data estÃ¡ vazio, logo os valores serÃ£o apresentados como NaN

```python
dfs = pd.DataFrame(
    data=[],
    columns=['LÃ³gica de ProgramaÃ§Ã£o', 'Python', 'Data Science', 'HTML & CSS', 'JavaScript', 'React Native'],
    index=['Amanda', 'Camila', 'Cinthia', 'Gisele', 'Helen', 'Carol', 'Clara', 'Paula']
)

dfs
```

Importaremos a biblioteca numpy - Numeric Python - para nos auxiliar a "popular" nosso DataFrame com as notas dentro do intervalo de 5 Ã  10 ao longo das colunas

```python
import numpy as np

dfs = dfs.apply(lambda nota: np.random.uniform(low=5.0, high=10.0, size=8), axis='index')
dfs
```

```python
#Inserir uma coluna para calcular a mÃ©dia das notas
# MÃ©todo mean
dfs['MÃ©dia'] = dfs.mean(axis='columns')
dfs
```
